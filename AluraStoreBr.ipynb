#1. Análise do faturamento


```python
def valor_total(loja):
    soma_preco = loja.Preço.sum()
    soma_frete = loja.Frete.sum()
    total = soma_preco + soma_frete
    return soma_preco, soma_frete, total
```

```python
valor_total(loja1)
```

```python
valor_total(loja2)
```

```python
valor_total(loja3)
```

```python
valor_total(loja4)
```

# 2. Vendas por Categoria


```python
def qtd_produtos_vendidos(loja):
    venda_produto = loja.groupby('Categoria do Produto')[['Produto']].count().sort_values('Produto')
    venda_total = loja.groupby('Categoria do Produto')[['Produto']].count().sum()
    print(f'{venda_produto}')
    print(f'Total de produtos vendidos: {venda_total}')
```

```python
qtd_produtos_vendidos(loja1)
```

```python
qtd_produtos_vendidos(loja2)
```

```python
qtd_produtos_vendidos(loja3)
```

```python
qtd_produtos_vendidos(loja4)
```

# 3. Média de Avaliação das Lojas

```python
def media_avaliacao_clientes(loja):
    return loja['Avaliação da compra'].mean()
```

# 4. Produtos Mais e Menos Vendidos

```python
def produtos_mais_menos_vendidos(loja):
    maior_venda = max(loja.groupby('Produto')[['Produto']].value_counts())
    menor_venda = min(loja.groupby('Produto')[['Produto']].value_counts())
    print(f'Maior venda {maior_venda}')
    print(f'Menor venda {menor_venda}')
    
```

```python
produtos_mais_menos_vendidos(loja1)
```

```python
produtos_mais_menos_vendidos(loja2)
```

```python
produtos_mais_menos_vendidos(loja3)
```

```python
produtos_mais_menos_vendidos(loja4)
```
